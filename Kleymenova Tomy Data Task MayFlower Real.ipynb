```python
import pandas as pd
import numpy as np

bankVarsData = pd.read_csv("bnk_vars.csv")
bankVarsData
```

```python
summaryofDeposits = pd.read_csv("sod.csv")
summaryofDeposits.sort_values(by=['RSSDID','BRNUM','YEAR'])

```

```python


"11.(Please use SAS to answer this question.)"
"a. Read in the data from the file “sod.csv,” find duplicates and drop them."
summaryofDeposits = summaryofDeposits.drop_duplicates(keep ='first', inplace = False)
summaryofDeposits.sort_values(by=['RSSDID','BRNUM','YEAR'])


```

```python
'b. Fill in the missing values in DEPSUMBR with the average of DEPSUMBR across'
"branches within the bank in the missing year."

summaryofDeposits['DEPSUMBR'] = summaryofDeposits.groupby(['RSSDID','BRNUM','YEAR'])['DEPSUMBR'].transform(lambda x: x.fillna(x.mean())) 

summaryofDeposits.sort_values(by=['RSSDID','BRNUM','YEAR'])
```

```python
'c. Add up DEPSUMBR at the bank level by year and rename it TOT_DEPOSITS.'
summaryofDeposits = summaryofDeposits.pivot_table(index=['RSSDID','BRNUM','YEAR'],values= ["DEPSUMBR"], aggfunc =np.sum)

summaryofDeposits

```

```python
summaryofDeposits.columns = [ 'TOT_DEPOSITS']
                             
summaryofDeposits
```

```python
bankVarsData['DATESTATA'] = bankVarsData['DATE'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
bankVarsData
```

```python
bankVarsData['DATEQTR'] = pd.PeriodIndex(pd.to_datetime(bankVarsData['DATESTATA']), freq='Q')
bankVarsData
```

```python
bankVarsData['DATE'] = bankVarsData['DATE'].astype(int)


bankVarsData['REGULATION'] = np.where(bankVarsData['DATE'] >= 19990701 , 1, 0)
bankVarsData
```

```python
"Please create a new variable TREATMENT by assigning one to banks that had an event"
"(EVENT = 1) at any point in their existence. Otherwise, assign zero."

eventData = bankVarsData[bankVarsData['event'] == 1]

bankVarsData['TREATMENT'] = np.where(bankVarsData['RSSDID'].isin(eventData['RSSDID']), 1, 0)
bankVarsData
```
